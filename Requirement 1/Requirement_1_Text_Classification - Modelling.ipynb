
# Modeling 

```python
#conterting to list
type(data_content["Content"])
Content = data_content["Content"].tolist()
cont = cont["Content"].tolist()
```

```python
y = y["Label"].tolist()
id_n = id_n["Id"].tolist()
```

TFIDF

```python
#tfidf for make the words to numerical
vectorizer = TfidfVectorizer(binary=True)
X = vectorizer.fit_transform(Content)
vectorizer.get_feature_names_out()
X.toarray()
```

```python
X_content_test = vectorizer.transform(cont)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

#fine tuning 
param_grid = [{'n_estimators':[100], 'max_depth':[None], 'max_features':["sqrt"]}]
forest = RandomForestClassifier()
grid_search = GridSearchCV(forest, param_grid, cv=5)
grid_search.fit(X_train, y_train)
final = grid_search.best_params_
print(final)
```

```python
from sklearn.metrics import accuracy_score
#Accuracy
y_pred = grid_search.predict(X_test)
print("Accuracy for Random Forest on CV data: ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
#classification report
print(classification_report(y_test, y_pred))
```

SVM

```python
from sklearn.svm import SVC

#fine tuning 
param_grid = [{'C':[1], 'gamma':["auto"], 'kernel':["rbf"]}]
# param_grid = [{'C':[0.1, 1, 10, 100, 1000], 'gamma':[1, 0.1, 0.01, 0.001, 0.0001], 'kernel':["rbf","linear"]}]
svc = SVC()
grid_search = GridSearchCV(svc, param_grid, cv=5)
grid_search.fit(X_train, y_train)
final = grid_search.best_params_
print(final)
```

```python
y_pred = grid_search.predict(X_test)
print("Accuracy for SVM on CV data: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

#fine tuning 
1e-9
param_grid = [{'var_smoothing': [1e-9]}]
# param_grid = [{'var_smoothing': np.logspace(0,-9, num=100)}]
NaiveBayes = GaussianNB()
grid_search = GridSearchCV(NaiveBayes, param_grid, cv=5)
X_train_NB = X_train.todense()
grid_search.fit(X_train_NB, y_train)
final = grid_search.best_params_
print(final)
```

```python
X_test_NB = X_test.todense()
y_pred = grid_search.predict(X_test_NB)
print("Accuracy for Naive Bayes on CV data: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = [{'n_neighbors':[5], 'weights':["uniform"], 'metric':["minkowski"]}]
# param_grid = [{'n_neighbors':[3,5,10] 'weights' : ['uniform','distance'],'metric' : ['minkowski','euclidean','manhattan']}
KNN = KNeighborsClassifier()
grid_search = GridSearchCV(KNN, param_grid, cv=5)
grid_search.fit(X_train, y_train)
final = grid_search.best_params_
print(final)
```

```python
y_pred = grid_search.predict(X_test)
print("Accuracy for KNN on CV data: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

XGBOOST

```python
from xgboost import XGBClassifier

param_grid = [{'alpha':[10], 'max_depth':[5], 'n_estimators':[10]}]
GB = XGBClassifier()
grid_search = GridSearchCV(GB, param_grid, cv=5)
grid_search.fit(X_train, y_train)
final = grid_search.best_params_
print(final)
```

```python
y_pred = grid_search.predict(X_test)
print("Accuracy for XGBOOST on CV data: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
